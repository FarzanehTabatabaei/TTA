```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv("breast_cancer.csv")

alt.Chart(df).mark_point().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = "breast-quad"
).properties(
    title = "Breast Cancer"
)
```

It is the first graph of the home learning task and for this task, I selected breast cancer data to categorise the size of tumore and the position of breast tumor based on age. By knowing the age and the size of position of tumor, this visualisation will help us to predict the size of tumor for future patients.

```python
alt.Chart(df).mark_bar().encode(
    x = alt.X("age") ,
    y = "count()"
)
```

As you can see, the vertical axis in this grapg shows the age of patients and the horizontal axis shows the count of recordes. I used the bar chart for this visualisation because it helps us to have better insight. The reason of choosing this bar chart is that it helps us to understand which age groups are more at risk(In this graph the women beween 50 to 59) and to start preventation from high-risk groups to reduce the disease.

```python
alt.Chart(df).mark_bar().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = "breast-quad"
).properties(
    title = "Breast Cancer"
)
```

This graph is like the first graph but we used the bar chart for having better insight.

```python
alt.Chart(df).mark_bar().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = "breast-quad" ,
    column = "breast-quad"
).properties(
    title = "Breast Cancer"
)
```

In this visualisation we split our graph to six different graphs based on the breast_quad column to have more clear insight. for example based on this visualisation, most patients have tumore in left-low and left-up position and also most of women who have left-low tumor are between the age of 40 to 49 and also the most women who have left-up tumor are between the age of 50 to 59 and 60 to 69. In future, if we have a 46 year-old patient, our initial prediction is that the tumor is in the left-low of her breast.

```python
interval = alt.selection_interval(encodings=["y"])

scatter = alt.Chart(df).mark_point().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = alt.condition(interval, "breast-quad", alt.value("lightgray"))
).properties(
    selection = interval
)

bar = alt.Chart(df).mark_bar().encode(
    x = alt.X("age") ,
    y = "count()"
)

scatter & bar
```

```python
interval = alt.selection_interval(encodings=["y"])

scatter = alt.Chart(df).mark_point().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = alt.condition(interval, "breast-quad", alt.value("lightgray"))
).properties(
    selection = interval
)

bar = alt.Chart(df).mark_bar().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = "tumor-size"
)

scatter | bar
```

In two previous graphs. we tried to combined to graphs with each other in two different ways(in the first way two graphs are next to each other and in the second way, one of graphs is above another one.)

```python
interval = alt.selection_interval(encodings=["y"])

scatter = alt.Chart(df).mark_point().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = alt.condition(interval, "breast-quad", alt.value("lightgray"))
).properties(
    selection = interval
)

bar = alt.Chart(df).mark_bar().encode(
    x = "age" ,
    y = "tumor-size" ,
    color = "tumor-size"
).transform_filter(
    interval
)

alt.vconcat(scatter, bar)
```

This graph is like two previous ones but in this chart, with moving the gray rectangle on the top chart, the buttom chart will change.
